```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.graph_objs as go
```

```python
# Tickers, Start and End Dates or (periods) intervals
# JPMC - JPM... 1day @ 1 min intervals
```

```python
jpmc_df = yf.download(tickers='JPM',period='1d',interval='1m')
```

```python
jpmc_df.head()
```

```python
jpmc_df['Middle Band'] = jpmc_df['Close'].rolling(window=21).mean()
jpmc_df['Upper Band'] = jpmc_df['Middle Band'] + 1.96*jpmc_df['Close'].rolling(window=21).std()
jpmc_df['Lower Band'] = jpmc_df['Middle Band'] - 1.96*jpmc_df['Close'].rolling(window=21).std()
```

```python
jpmc_df.tail()
```

```python
# Use Plotly to Graph
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(x=jpmc_df.index, y = jpmc_df['Middle Band'], line=dict(color='blue',width=.7),name='Middle Band'))
fig.add_trace(go.Scatter(x=jpmc_df.index, y = jpmc_df['Upper Band'], line=dict(color='red',width=1.5),name='Upper Band (Sell/$ DOWN)'))
fig.add_trace(go.Scatter(x=jpmc_df.index, y = jpmc_df['Lower Band'], line=dict(color='green',width=1.5),name='Lower Band (Buy/$ UP'))
```

```python
#Candlestick
fig.add_trace(go.Candlestick(x=jpmc_df.index,
                open=jpmc_df['Open'],
                high=jpmc_df['High'],
                low=jpmc_df['Low'],
                close=jpmc_df['Close'], name = 'market data'))
```

```python
# Add Meta data to the chart
fig.update_layout(title='Asset Price Evolution', yaxis_title='Price')
#X Axis
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Show
fig.show()
```
