```python
import numpy as np
import pandas as pd

# Data Source 
import yfinance as yf

# Visualizations
import plotly.graph_objs as go
```

```python
euro_usd = yf.download(tickers='EURUSD=X',period='1d',interval='15m')
euro_usd
rub_usd = yf.download(tickers='RUB=X',period='1d',interval='15m')
```

```python
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=rub_usd.index,
                open=rub_usd['Open'],
                high=rub_usd['High'],
                low=rub_usd['Low'],
                close=rub_usd['Close'], name = 'market data'))

# Add titles
fig.update_layout(
    title='USD to Rubles')

#Show
fig.show()
```
